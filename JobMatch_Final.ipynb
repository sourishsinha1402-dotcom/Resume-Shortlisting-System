```python
import os
import pdfplumber
import docx2txt
from docx import Document
import PyPDF2
import pandas as pd
from sentence_transformers import SentenceTransformer, util
```

Helpers for Reading Files

```python
def read_txt(path):
    with open(path, "r", encoding="utf-8", errors="ignore") as f:
        return f.read()

def read_docx(path):
    doc = docx.Document(path)
    return "\n".join([para.text for para in doc.paragraphs])

def read_pdf(path):
    text = ""
    with open(path, "rb") as f:
        reader = PyPDF2.PdfReader(f)
        for page in reader.pages:
            text += page.extract_text() or ""
    return text

def extract_text(path):
    ext = os.path.splitext(path)[1].lower()
    if ext == ".txt":
        return read_txt(path)
    elif ext == ".docx":
        return read_docx(path)
    elif ext == ".pdf":
        return read_pdf(path)
    else:
        raise ValueError(f"Unsupported file format: {ext}")
```

Read Files

```python
from google.colab import files
import os, shutil

# Clean old folders
for folder in ["resumes", "jobs"]:
    if os.path.exists(folder):
        shutil.rmtree(folder)
    os.makedirs(folder)

print("Upload Resumes (PDF, DOCX, TXT):")
uploaded_resumes = files.upload()
for fn in uploaded_resumes.keys():
    shutil.move(fn, os.path.join("resumes", fn))

print("Upload Job Descriptions (PDF, DOCX, TXT):")
uploaded_jobs = files.upload()
for fn in uploaded_jobs.keys():
    shutil.move(fn, os.path.join("jobs", fn))
```

Read Files

```python
def read_file(filepath):
    text = ""
    ext = filepath.lower().split(".")[-1]

    if ext == "txt":
        with open(filepath, "r", encoding="utf-8", errors="ignore") as f:
            text = f.read()

    elif ext == "docx":
        doc = Document(filepath)
        text = "\n".join([para.text for para in doc.paragraphs])

    elif ext == "pdf":
        with open(filepath, "rb") as f:
            reader = PyPDF2.PdfReader(f)
            for page in reader.pages:
                text += page.extract_text() or ""

    return text.strip()


def load_files_from_folder(folder):
    texts, filenames = [], []
    for fname in os.listdir(folder):
        fpath = os.path.join(folder, fname)
        try:
            text = read_file(fpath)
            if text:
                texts.append(text)
                filenames.append(fname)
        except Exception as e:
            print(f"Could not read {fname}: {e}")
    return filenames, texts
```

Model Setup

```python
model = SentenceTransformer("multi-qa-mpnet-base-dot-v1")
```

Load Data

```python
resume_names, resume_texts = load_files_from_folder("resumes")
job_names, job_texts = load_files_from_folder("jobs")

print(f"Loaded {len(resume_names)} resumes and {len(job_names)} jobs")
```

Compute Similarity

```python
resume_emb = model.encode(resume_texts, convert_to_tensor=True, show_progress_bar=True)
job_emb = model.encode(job_texts, convert_to_tensor=True, show_progress_bar=True)

similarity_matrix = util.cos_sim(resume_emb, job_emb).cpu().numpy()
```

Results

```python
rows = []
for j, jname in enumerate(job_names):
    job_scores = []
    for i, rname in enumerate(resume_names):
        job_scores.append({
            "Job": jname,
            "Resume": rname,
            "Similarity": round(float(similarity_matrix[i][j]), 4)
        })

    # Sort resumes by similarity (descending) and take top 10
    top_resumes = sorted(job_scores, key=lambda x: x["Similarity"], reverse=True)[:10]
    rows.extend(top_resumes)

# Convert to DataFrame
df_results = pd.DataFrame(rows)

output_file = "top10_resume_per_job.csv"
df_results.to_csv(output_file, index=False)

print(f"Top 10 resumes per job saved to {output_file}")
df_results.head(20)
```

#Evaluation Metrics

```python
import numpy as np
from sklearn.metrics import silhouette_score, davies_bouldin_score
```

```python
print("\n=== Overall Similarity Stats ===")
print(f"Mean: {np.mean(similarity_matrix):.4f}")
print(f"Median: {np.median(similarity_matrix):.4f}")
print(f"Min: {np.min(similarity_matrix):.4f}")
print(f"Max: {np.max(similarity_matrix):.4f}")
print(f"Std Dev: {np.std(similarity_matrix):.4f}")

print("\n=== Per-Job Similarity Stats (first 5 jobs) ===")
for j, jname in enumerate(job_names[:5]):
    sims = similarity_matrix[:, j]
    print(f"{jname} → Mean={np.mean(sims):.4f}, Median={np.median(sims):.4f}, Max={np.max(sims):.4f}")
```

Clustering Quality of embeddings

```python
all_embeddings = np.vstack([resume_emb.cpu().numpy(), job_emb.cpu().numpy()])
labels = [0]*len(resume_names) + [1]*len(job_names)  # 0=resumes, 1=jobs

try:
    sil_score = silhouette_score(all_embeddings, labels)
    db_score = davies_bouldin_score(all_embeddings, labels)
    print("\n=== Clustering Metrics ===")
    print(f"Silhouette Score: {sil_score:.4f} (higher is better, range -1 to 1)")
    print(f"Davies-Bouldin Index: {db_score:.4f} (lower is better)")
except Exception as e:
    print(f"Clustering metrics skipped: {e}")
```
